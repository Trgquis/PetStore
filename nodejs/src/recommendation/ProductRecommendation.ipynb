```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
file_path = "./reviews.csv"

df = pd.read_csv(file_path)
df

```

```python
df = df[['user_id', 'product_id', 'score']]
df
```

```python
import pandas as pd

pivot_df = df.pivot_table(index='user_id', columns='product_id', values='score', fill_value=0)
print(pivot_df)
```

```python
item = list(pivot_df.columns)
user = list(pivot_df.index)
```

```python
import numpy as np
import pandas as pd

def random_fill(x):
    if pd.isnull(x):
        return np.random.randint(0, 6)  # Trả về một số ngẫu nhiên từ 0 đến 5 nếu giá trị là NaN
    else:
        return np.random.randint(0, 6)  # Trả về một số ngẫu nhiên từ 0 đến 5 nếu giá trị là NaN


data_matrix = pivot_df.applymap(random_fill).values

print(data_matrix)
```

```python
data_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

Sim = cosine_similarity(data_matrix,data_matrix)
Sim = np.round(Sim, decimals = 2)
Sim
```

```python
pd.DataFrame(Sim, columns=user, index=user)
```

```python
def predict(A, u,i, k):
    user_rated_i = np.where(A[:,i] !=0)[0]
    sim = Sim[u, user_rated_i]
    a = (np.argsort(sim))[-k:]
    nearest_s =sim[a]
    rating = A[user_rated_i[a],i]
    r_bar = (rating*nearest_s).sum()/(np.abs(nearest_s).sum())
    return np.round(r_bar)
```

```python
predict_matrix = data_matrix.copy()
for u in range(0, predict_matrix.shape[0]):
  for i in range(0, predict_matrix.shape[1]):
    if data_matrix[u,i] ==0:
      id = i
      predict_matrix[u,i] = predict(data_matrix,u,id, 25)
predict_matrix
```

```python
import numpy as np

def predict_item_average(A, i):
    rated_items = A[:, i]
    rated_items = rated_items[rated_items != 0]
    if len(rated_items) == 0:
        return np.nan
    return np.mean(rated_items)

predicted_rating = predict_item_average(data_matrix, i)
print(predicted_rating)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data_matrix, test_size=0.3, random_state=50)

def calculate_rmse_user_average(predicted_matrix, true_matrix):
    predicted_values = predicted_matrix[true_matrix != 0]
    true_values = true_matrix[true_matrix != 0]
    return np.sqrt(mean_squared_error(true_values, predicted_values))

# Tính dự đoán cho tất cả các mục trong tập kiểm thử
predicted_matrix = np.zeros_like(test_data)
for u in range(test_data.shape[0]):
    for i in range(test_data.shape[1]):
        if test_data[u, i] == 0:  # Chỉ dự đoán cho các mục chưa được đánh giá
            predicted_matrix[u, i] = predict(train_data, u, i, 25)  # Sử dụng hàm predict để dự đoán

# Tính toán RMSE
rmse = calculate_rmse_user_average(predicted_matrix, test_data)
print("RMSE User-KNNCF:", rmse)
```

```python
train_data2, test_data2 = train_test_split(data_matrix, test_size=0.3, random_state=42)

def calculate_rmse_item_average(predicted_matrix, true_matrix):
    predicted_values = predicted_matrix[true_matrix != 0]
    true_values = true_matrix[true_matrix != 0]
    return np.sqrt(mean_squared_error(true_values, predicted_values))

# Tính dự đoán trung bình cho mỗi mục trong tập kiểm thử
predicted_matrix_item_average = np.zeros_like(test_data2)
for i in range(test_data2.shape[1]):
    for u in range(test_data2.shape[0]):
        if test_data2[u, i] == 0:  # Chỉ dự đoán cho các mục chưa được đánh giá
            predicted_matrix_item_average[u, i] = predict_item_average(train_data2, i)  # Sử dụng hàm predict_item_average để dự đoán

# Tính toán RMSE
rmse_item_average = calculate_rmse_item_average(predicted_matrix_item_average, test_data2)
print("RMSE của Item Average:", rmse_item_average)
```

```python
import pandas as pd

# Tạo DataFrame chứa dữ liệu RMSE
rs = {
    "Method": ["User-KNNCF", "Item Average"],
    "RMSE": [rmse, rmse_item_average]
}

dfrs = pd.DataFrame(rs)

# In ra bảng so sánh
print(dfrs)
```
